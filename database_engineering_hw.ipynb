```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
```

```python
hi_measurement = pd.read_csv('clean_hawaii_measurements.csv')
hi_stations = pd.read_csv('hawaii_stations.csv')
```

```python
hi_measurement_df = pd.DataFrame(hi_measurement)
hi_measurement_df.columns = ['ID','station','date','prcp','tobs']
hi_measurement_df.head()
```

```python
hi_stations_df = pd.DataFrame(hi_stations)
hi_stations_df
```

```python
Base = declarative_base()
```

```python
class Measurement(Base):
    __tablename__ = 'measurement'
    id = Column(Float, primary_key=True)
    station = Column(String)
    date = Column(String)
    prcp = Column(Float)
    tobs = Column(Float)
```

```python
class Station(Base):
    __tablename__ = 'station'
    station = Column(String(255), primary_key=True)
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
engine = create_engine('sqlite:///hawaii_engine.sqlite')
```

```python
Base.metadata.create_all(engine)
```

```python
session = Session(engine)
```

```python
#session.query(Measurement).delete()
```

```python
day = 0
for row in hi_measurement_df.iterrows():
    session.add(Measurement(id = hi_measurement_df['ID'][day], 
                            station= hi_measurement_df['station'][day], 
                            date = hi_measurement_df['date'][day], 
                            prcp = hi_measurement_df['prcp'][day], 
                            tobs = hi_measurement_df['tobs'][day]))
    
    day = day + 1
```

```python
#session.rollback()
```

```python
session.commit()
```

```python
items = session.query(Measurement)
for item in items:
    print('-'*12)
    print(f'station: {item.station}')
    print(f'date: {item.date}')
    print(f'prcp: {item.prcp}')
    print(f'tobs: {item.tobs}')
```

```python
day_station = 0
for stations in hi_stations_df.iterrows():
    session.add(Station(station= hi_stations_df['station'][day_station], 
                        name = hi_stations_df['name'][day_station], 
                        latitude = hi_stations_df['latitude'][day_station], 
                        longitude = hi_stations_df['longitude'][day_station],
                        elevation = hi_stations_df['elevation'][day_station]))
    
    day_station = day_station + 1
```

```python
session.commit()
```

```python
station_items = session.query(Station)
for station_data in station_items:
    print('-'*12)
    print(f'station: {station_data.station}')
    print(f'name: {station_data.name}')
    print(f'latitude: {station_data.latitude}')
    print(f'longitude: {station_data.longitude}')
    print(f'elevation: {station_data.elevation}')
```
